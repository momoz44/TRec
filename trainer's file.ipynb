```python
uid = []
username = []
trainer_name = []
trainer_email = []
trainer_status = []
score = []
language = []
location = []
cv = []
privided_categories = []
provided_courses = []
```

```python
for tr in trainers:
    uid.append(tr['uid'])
    username.append(tr['username'])
    trainer_name.append(tr['trainer_name'])
    trainer_email.append(tr['trainer_email'])
    trainer_status.append(tr['trainer_status'])
    score.append(tr['score'])
    language.append(tr['language'])
    location.append(tr['location'])
    cv.append(tr['cv'])
    cates = []
    for j in tr['privided_categories']:
        cates.append(j['name'])
    privided_categories.append(cates)
    courses = []
    for j in tr['provided_courses']:
        courses.append(j['nid'])
    provided_courses.append(courses)
```

```python
result = pd.DataFrame({'uid':uid,'username':username,'trainer_name':trainer_name,'trainer_email':trainer_email,'trainer_status':trainer_status,'score':score,'language':language,'location':location,'privided_categories':privided_categories,'provided_courses':provided_courses})
```

```python
result.to_csv('inner_trainers0.csv', encoding="utf-8")
```
